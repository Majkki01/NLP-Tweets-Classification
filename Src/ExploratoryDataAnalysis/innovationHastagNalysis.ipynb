```python
import pandas as pd


def lowercase(data):
    data['Text'] = data['Text'].str.lower()
    return data
```

```python
df = pd.read_parquet('../../dataset/dataset.parquet')
df = lowercase(df)
```

```python
var = df[df['Text'].str.contains("#innovation")]
```

```python
import matplotlib.pyplot as plt
new_df = df.groupby(["Label"]).count().reset_index()

new_df

new_df.plot.bar(x="Label", y="Text")
plt.savefig('innovationdistribution.png')
plt.show()
```
