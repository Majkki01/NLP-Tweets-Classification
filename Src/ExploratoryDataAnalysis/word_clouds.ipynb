```python
import pandas as pd
import matplotlib.pyplot as plt
from wordcloud import WordCloud

df = pd.read_parquet('../../dataset/dataset_clean.parquet')
```

```python
cloud_positive=WordCloud(
                          background_color='black',
                          width=2000,
                          height=1600
                         ).generate(" ".join(df[df['Label']=='positive']['Text']))
cloud_negative=WordCloud(
                          background_color='black',
                          width=2000,
                          height=1600
                         ).generate(" ".join(df[df['Label']=='negative']['Text']))
cloud_litigious=WordCloud(
                          background_color='black',
                          width=2000,
                          height=1600
                         ).generate(" ".join(df[df['Label']=='litigious']['Text']))
cloud_uncertainty=WordCloud(
                          background_color='black',
                          width=2000,
                          height=1600
                         ).generate(" ".join(df[df['Label']=='uncertainty']['Text']))
```

```python
plt.imshow(cloud_positive)
plt.title('Positive Tweets',fontsize=40)
plt.axis('off')
plt.savefig('positive_cloud.png')
plt.show()

plt.imshow(cloud_negative)
plt.title('Negative Tweets',fontsize=40)
plt.axis('off')
plt.savefig('negative_cloud.png')
plt.show()

plt.imshow(cloud_litigious)
plt.title('Litigious Tweets',fontsize=40)
plt.axis('off')
plt.savefig('litigious_cloud.png')
plt.show()

plt.imshow(cloud_uncertainty)
plt.title('Uncertainty Tweets',fontsize=40)
plt.axis('off')
plt.savefig('uncertainty_cloud.png')
plt.show()
```
